```python
# Practicing Oops
class car:
    def __init__(self, brand, colour, speed):
        self.brand = brand
        self.colour = colour
        self.speed = speed
    
    def accelerate(self, amount):
        self.speed += amount
        print(f"{self.brand} accelareted to {self.speed} km/h.")

    def brake(self):
        self.speed = 0
        print(f"{self.brand} has stopped.")
```
